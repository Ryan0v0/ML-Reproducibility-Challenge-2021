```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(columns=['graph_idx', 'Method', 'Name', 'odd'])
data = pd.read_csv("log/syn1_top6/log_odd.csv", header=None, names=["graph_idx", "our_ground_truth_odd", "gnnexp_ground_truth_odd", "our_pred_label_odd", "gnnexp_pred_label_odd"])
for i, line in data.iterrows():
    df = df.append({'graph_idx': line['graph_idx'], 'Method': 'Gem', 'Dataset': 'MUTAG', 'Log-odds difference': line['our_pred_label_odd']}, ignore_index=True)
    df = df.append({'graph_idx': line['graph_idx'], 'Method': 'GNNExplainer', 'Dataset': 'MUTAG', 'Log-odds difference': line['gnnexp_pred_label_odd']}, ignore_index=True)

for i, logodd in enumerate(np.load("log/violin_rawdata/syn1_log_odd.npy")):
    df = df.append({'graph_idx': i, 'Method': 'PGExplainer', 'Dataset': 'MUTAG', 'Log-odds difference': logodd}, ignore_index=True)
    
sns.set(style="whitegrid", font_scale=2)
fig, ax = plt.subplots(figsize=(10,8))
ax = sns.violinplot(x='Method', y='Log-odds difference', data=df, split=True, scale="count", inner="quartile")  
plt.tight_layout()
sns.despine(left=True)
plt.savefig("BA-Shape_log_odd.pdf")
plt.show()
```
